# 1. What are the top 10 PROSPECTOR?

```python
PROSPECTOR_percentage = new_df['PROSPECTOR'].value_counts() * 100/len(new_df)
sns.set(style="darkgrid")

# Days_count
plt.figure(figsize=(20,15))
plt.yticks(fontsize=20)
sns.barplot(PROSPECTOR_percentage.values[:10], PROSPECTOR_percentage.index[:10])

plt.title('Top 10 PROSPECTOR',fontsize=30)
plt.ylabel('Types of PROSPECTOR', fontsize=20)
plt.xlabel('Percentage of PROSPECTOR',fontsize=20)
plt.show();
```

# Insights from the above chart

From the chart, we can see the higher RO section  is contributing nearly 10% which is the highest followed  and AB which is nearly 7%

# 1. What are the top 15 Batch ?

```python
batch_percentage = new_df['Batch Name'].value_counts() * 100/len(new_df)
sns.set(style="darkgrid")

# Batch_count
plt.figure(figsize=(30,15))
plt.yticks(fontsize=20)
sns.barplot(batch_percentage.values[:20], batch_percentage.index[:20])

plt.title('Top 15 Batch',fontsize=30)
plt.ylabel('Batch', fontsize=20)
plt.xlabel('Percentage of Batch',fontsize=20)

plt.show()
```
